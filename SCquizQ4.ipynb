## Get the A matrix along [322] in triclinic

```python
#uvw is uvw prime and first second third are the basis vectors uvw
uvw = np.array([3,2,2])
first = np.array([1,0,0])
second = np.array([0,1,0])
third = np.array([0,0,1])


a = 4.75
b = 9.32
c = 11.65
alpha = np.radians(95.07)
beta = np.radians(95.31)
gamma = np.radians(89.6)

#CsS3 is triclinic
Gmatrix = np.array([[a**2, a*b*np.cos(gamma), a*c*np.cos(beta)], 
                    [b*a*np.cos(gamma), b**2, b*c*np.cos(alpha)], 
                    [c*a*np.cos(beta), c*b*np.cos(alpha), c**2]])


a11 = np.dot(np.dot(uvw, Gmatrix), first.T)
a11 = a11/(np.sqrt(np.dot(np.dot(uvw, Gmatrix), uvw.T)) * np.sqrt(np.dot(np.dot(first, Gmatrix), first.T)))

a12 = np.dot(np.dot(uvw, Gmatrix), second.T)
a12 = a12/(np.sqrt(np.dot(np.dot(uvw, Gmatrix), uvw.T)) * np.sqrt(np.dot(np.dot(second, Gmatrix), second.T)))

a13 = np.dot(np.dot(uvw, Gmatrix), third.T)
a13 = a13/(np.sqrt(np.dot(np.dot(uvw, Gmatrix), uvw.T)) * np.sqrt(np.dot(np.dot(third, Gmatrix), third.T)))

Amatrix = np.array([a11, a12, a13])

print('Amatrix: \n', Amatrix)
```

## Solve for dielectric permmitivty k'


```python
#Given k tensor
k = np.array([[6.692, 0.893, 0.1519],
               [0.893, 11.4, 0.1739],
               [0.1519, 0.1739, 6.723]])

print('k tensor: \n', k)
```

```python
kprime = np.dot(np.dot(Amatrix, k), Amatrix.T)

print('k\' tensor in [322]: \n', kprime)
```
